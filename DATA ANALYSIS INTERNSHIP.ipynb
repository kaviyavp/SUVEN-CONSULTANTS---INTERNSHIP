```python
Data = Data.drop(['Loud Cover'], axis = 1)
```

```python
Data['Formatted Date'] = pd.to_datetime(Data['Formatted Date'],
utc = True)
```

```python
Data = Data.set_index("Formatted Date") 
Data
```

```python
df_column = ['Apparent Temperature (C)', 'Humidity']
df_monthly_mean = Data[df_column].resample("MS").mean()
df_monthly_mean.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_monthly_mean['Humidity'], label = 'Humidity', color = 'red', linestyle='dashed')
plt.plot(df_monthly_mean['Apparent Temperature (C)'], label = 'App temp', color = 'blue')
plt.title('Variation in Apparent Temperature v/s Humidity', fontsize = 25);
plt.legend(loc =0, fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 13)
```

```python
df1 = df_monthly_mean[df_monthly_mean.index.month==1]
print(df1)

df1.dtypes
```

```python
import matplotlib.dates as mdates
from datetime import datetime 

fig, ax = plt.subplots(figsize=(15,5))
ax.plot(df1.loc['2006-01-01':'2016-01-01', 'Apparent Temperature (C)'], marker='o', linestyle='-',label='Apparent Temperature (C)')
ax.plot(df1.loc['2006-01-01':'2016-01-01 ', 'Humidity'], marker='o', linestyle='-',label='Humidity')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.legend(loc = 'center right')
ax.set_xlabel('Month of January')
```
